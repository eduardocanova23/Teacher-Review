```python
def get_course_info():
    data = []
    for cur in cursos:
        name, link = cur
        course_response = requests.get(f'https://siga.ufrj.br/sira/gradeHoraria/{link}')
        course_soup = BeautifulSoup(course_response.text, 'html.parser')
        table_rows = course_soup.find_all('tr', class_=['tableBody1', 'tableBody2'])
        for row in table_rows:
            cells = row.find_all('td')
            if len(cells) > 0:
                codigo = cells[0].text.strip()
                if codigo == 'Código':
                    continue
                turma = cells[2].text.strip()
                professor = cells[5].text.strip()
                if codigo and professor and turma:
                    data.append([codigo, turma, professor, name])
        print(f'Complete course: {name}')
    return data
```

## Create Scraped DataFrame

```python
siga_info = pd.DataFrame(scraped_data, columns=['cod', 'nome_abv', 'professor', 'curso'])
```

group by longest common substring

ex: convert Calc 1 [AB] and Calc 1 [NOITE] to Calc 1

```python
def long_substr(data):
    substr = ''
    if len(data) > 1 and len(data[0]) > 0:
        for i in range(len(data[0])):
            for j in range(len(data[0])-i+1):
                if j > len(substr) and all(data[0][i:i+j] in x for x in data):
                    substr = data[0][i:i+j]
        return substr
    else:
        return data[0]
```

```python
siga_info_by_cod = siga_info.groupby('cod')['nome_abv'].agg(lambda x: long_substr(x.to_list()))
siga_info_by_cod = siga_info_by_cod.reset_index()
siga_info_by_prof = siga_info.loc[:, ['cod', 'professor']].drop_duplicates()
siga_result = siga_info_by_cod.merge(siga_info_by_prof, on="cod", how="inner")
```

## Cleaning dataframe

- Clean Não Informado

```python
siga_result = siga_result[siga_result['professor'] != 'Não Informado']
```

- Clean subjects with string ALUNO

```python
siga_result.loc[:, 'nome_abv'] = siga_result.loc[:, 'nome_abv'].str.lower().str.split('aluno', expand=True)[0]
siga_result.loc[:, 'nome_abv'] = siga_result.loc[:, 'nome_abv'].str.lower().str.split('alun', expand=True)[0]
```

- remove aditional \<space>

```python
siga_result.loc[:, 'nome_abv']  = siga_result.loc[:, 'nome_abv'].str.strip()
siga_result.loc[:, 'cod']  = siga_result.loc[:, 'cod'].str.strip()
siga_result.loc[:, 'professor']  = siga_result.loc[:, 'professor'].str.strip()
```

- drop duplicates

```python
siga_result = siga_result.drop_duplicates()
```
